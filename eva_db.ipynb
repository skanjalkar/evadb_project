```python
cursor.query("DROP TABLE IF EXISTS MyPDFs").df()
# cursor.query("""CREATE TABLE IF NOT EXISTS MyPDFs(
#   id UNIQUE
# )""").df()
cursor.query("LOAD PDF 'john_doe.pdf' INTO MyPDFs").df()
```

```python
cursor.query("SELECT * FROM MyPDFs").df()
```

```python
data = cursor.query("""
    SELECT data
    FROM MyPDFs
""").df()
print(data)
```

```python
list_data = data['mypdfs.data'].tolist()
print(list_data)
```

```python
cursor.query("""
    CREATE FUNCTION IF NOT EXISTS TextClassifier
    TYPE HuggingFace
    TASK 'text-classification'
    MODEL 'distilbert-base-uncased-finetuned-sst-2-english'
""").df()
```

```python
cursor.query("""
    CREATE FUNCTION IF NOT EXISTS TextSummarizer
    TYPE HuggingFace
    TASK 'summarization'
    MODEL 'facebook/bart-large-cnn'
""").df()
```

```python
cursor.query("""
    SELECT data, TextSummarizer(data)
    FROM MyPDFs
    WHERE page = 1 AND paragraph >= 1 AND paragraph <= 3 AND TextClassifier(data).label = 'NEGATIVE'
""").df()
```
